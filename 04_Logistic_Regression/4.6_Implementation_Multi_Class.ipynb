````python
# creation of the data set
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=1000, n_features=5, n_classes=3, random_state=42, n_informative=3)
````

Error was:ValueError: n_classes(3) * n_clusters_per_class(2) must be smaller or equal 2**n_informative(2)=4

#### What the error says

```text
n_classes(3) * n_clusters_per_class(2) <= 2^n_informative(2)
```

👉 Plug in your values:

* Classes = 3
* Clusters per class = 2
  👉 Total clusters = **3 × 2 = 6**

Right side:

* Informative features = 2
  👉 Possible separations = **2² = 4**

---

What it means to say is, we are trying to create 6 possible groups of the data but n_informative was set to 2
so that 2^2 only 4 separate groups are possible.

#### Problem

👉 You are trying to create **6 clusters**
👉 But your feature space can only support **4 separations**

👉 So mathematically → **impossible to separate them properly**

---

#### How to fix it

You have 3 options:

---

#### Option 1: Increase informative features (BEST fix)

```python
n_informative = 3
```

Now:

* 2³ = 8 ≥ 6 ✔️

---

#### Option 2: Reduce clusters per class

```python
n_clusters_per_class = 1
```

Now:

* 3 × 1 = 3 ≤ 4 ✔️

---


````python
# split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.33)
````

````python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression

regressor = LogisticRegression()
classifier = OneVsRestClassifier(regressor, n_jobs=-1)

classifier.fit(x_train, y_train)
````

````python
y_pred = classifier.predict(x_test)
````

````python
classifier.predict_proba(x_test).round(2)
````
