#### First — What each column means 

* **Precision** → “When model says this class, how often is it right?”
* **Recall** → “Out of actual items of this class, how many did it catch?”
* **F1-score** → Balance between precision & recall
* **Support** → Number of actual samples

---

#### Now report (real interpretation)

#### Class 0

* Precision = **0.89**
* Recall = **0.81**
* F1 = **0.85**
* Support = **172**

👉 Meaning:

* When model predicts **0**, it's correct **89% of the time** → strong precision
* But it only catches **81% of actual 0s**

👉 You're **missing some class 0 cases** (false negatives for class 0)

---

#### Class 1

* Precision = **0.81**
* Recall = **0.89**
* F1 = **0.85**
* Support = **158**

👉 Meaning:

* When model predicts **1**, it's correct **81% of the time**
* But it catches **89% of actual 1s**

👉 You are **over-predicting class 1 a bit** (more false positives)

---

#### Big Pattern (THIS is important)

Look carefully:

| Class | Precision | Recall |
| ----- | --------- | ------ |
| 0     | HIGH      | LOWER  |
| 1     | LOWER     | HIGH   |

👉 This means:

* Model is slightly **biased toward predicting class 1**
* It prefers catching class 1 even if it makes some mistakes

---

#### Accuracy = 0.85

👉 85% of total predictions are correct

### Hyper parameter tuning

1) Grid Search CV

```python
# Prepare the parameters
model = LogisticRegression()

# parameters
params = [
    # lbfgs + sag → only l2
    {
        "solver": ["lbfgs", "sag"],
        "penalty": ["l2"],
        "C": [0.01, 0.1, 1, 10]
    },

    # liblinear → l1, l2
    {
        "solver": ["liblinear"],
        "penalty": ["l1", "l2"],
        "C": [0.01, 0.1, 1, 10]
    },

    # saga → everything
    {
        "solver": ["saga"],
        "penalty": ["l1", "l2"],
        "C": [0.01, 0.1, 1, 10]
    },

    # elasticnet ONLY with saga
    {
        "solver": ["saga"],
        "penalty": ["elasticnet"],
        "C": [0.01, 0.1, 1, 10],
        "l1_ratio": [0.3, 0.5, 0.7]
    }
]

# params
params
```

```python
# getting the stratified cvs
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(n_splits=5)
```

```python
# handling the grid search cv
from sklearn.model_selection import GridSearchCV
search = GridSearchCV(estimator=model, param_grid=params, scoring="accuracy", n_jobs=-1, cv=cv)
search.fit(x_train, y_train)
```

```python
search.best_score_
```

```python
search.best_params_
```

```python
y_pred = search.predict(x_test)
y_pred
```

GridSearchCV is computationally expensive because it performs an exhaustive search over all possible combinations of hyperparameters.
If you have k parameters and each has n values, total combinations = nᵏ.
With cross-validation (cv folds), total model trainings = nᵏ × cv. \
👉 This grows exponentially and becomes very slow as parameters increase.

RandomizedSearchCV, on the other hand, samples a fixed number of random combinations from the parameter space. \
Instead of trying all combinations, it tries only n_iter × cv trainings. \
👉 This makes it much faster and scalable, especially for large search spaces.

```python
model = LogisticRegression()

params = [
    # lbfgs + sag → only l2
    {
        "solver": ["lbfgs", "sag"],
        "penalty": ["l2"],
        "C": [0.01, 0.1, 1, 10]
    },

    # liblinear → l1, l2
    {
        "solver": ["liblinear"],
        "penalty": ["l1", "l2"],
        "C": [0.01, 0.1, 1, 10]
    },

    # saga → everything
    {
        "solver": ["saga"],
        "penalty": ["l1", "l2"],
        "C": [0.01, 0.1, 1, 10]
    },

    # elasticnet ONLY with saga
    {
        "solver": ["saga"],
        "penalty": ["elasticnet"],
        "C": [0.01, 0.1, 1, 10],
        "l1_ratio": [0.3, 0.5, 0.7]
    }
]
```

```python
from sklearn.model_selection import RandomizedSearchCV
search = RandomizedSearchCV(estimator=model, param_distributions=params,scoring="accuracy",n_jobs=-1,cv=cv)

search.fit(x_train, y_train)
```
