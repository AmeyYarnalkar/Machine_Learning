## Holiday Package Prediciton

### 1) Problem statement.
"Trips & Travel.Com" company wants to enable and establish a viable business model to expand the customer base.
One of the ways to expand the customer base is to introduce a new offering of packages. Currently, there are 5 types of packages the company is offering * Basic, Standard, Deluxe, Super Deluxe, King. Looking at the data of the last year, we observed that 18% of the customers purchased the packages. However, the marketing cost was quite high because customers were contacted at random without looking at the available information.
The company is now planning to launch a new product i.e. Wellness Tourism Package. Wellness Tourism is defined as Travel that allows the traveler to maintain, enhance or kick-start a healthy lifestyle, and support or increase one's sense of well-being.
However, this time company wants to harness the available data of existing and potential customers to make the marketing expenditure more efficient.
### 2) Data Collection.
The Dataset is collected from https://www.kaggle.com/datasets/susant4learning/holiday-package-purchase-prediction
The data consists of 20 column and 4888 rows.

#### Data Cleaning


```python
# Understanding the data
print(df.shape)

# checking the data type
df.info()
```

```python
# checking the duplicate rows
df.duplicated().sum()
```

```python
# checking for the duplicate columns
df.T.duplicated().sum()
```

```python
# handling the missing values
df.isna().sum()
```

```python
# handling the column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
df
```

```python
# filling the missing values

#age
df["age"] = df["age"].fillna(df["age"].median())

#type of contact
df["typeofcontact"] = df["typeofcontact"].astype("category").fillna(df["typeofcontact"].mode())

#duration of the pitch
df["durationofpitch"] = df["durationofpitch"].fillna(df["durationofpitch"].mean())

# number of follow ups
df["numberoffollowups"] = df["numberoffollowups"].fillna(df["numberoffollowups"].mean())

# prefered property star
df["preferredpropertystar"] = df["preferredpropertystar"].dropna()

#number of the trips
df["numberoftrips"] = df["numberoftrips"].fillna(df["numberoftrips"].mean())

#number of children visiting 
df["numberofchildrenvisiting"] = df["numberofchildrenvisiting"].fillna(df["numberofchildrenvisiting"].mean())

# monthly income
df["monthlyincome"] = df["monthlyincome"].fillna(df["monthlyincome"].mean())
```

```python
# remove the remaining the missing values
df = df.dropna()
df.isna().sum()
```

```python
# handling the data types
df.info()
```

```python
# handling the categorical
def convertToCategory():
    cols = ["occupation","gender","productpitched","maritalstatus","designation"]
    
    for col in cols:
        df[col] = df[col].astype("category")

convertToCategory()
    
```

```python
df.info()
```

```python
# handling the extreme values and outliers
df["gender"] = df["gender"].str.replace("Fe Male","Female")
df["maritalstatus"] = df["maritalstatus"].str.replace("Single","Unmarried")
```

```python
# Remove the unwanted column
df.drop("customerid", axis=1, inplace=True)
```

#### Feature Extraction

```python
# total visiting
df["totalvising"] = df["numberofchildrenvisiting"] + df["numberofpersonvisiting"]
df.drop(columns=["numberofchildrenvisiting","numberofpersonvisiting"], axis=1, inplace=True)
```

```python
def feature_types(df):

    # numerical features
    numerical_features = df.select_dtypes(
        include=['int64', 'float64']
    ).columns

    # categorical features
    categorical_features = df.select_dtypes(
        include=['object', 'category']
    ).columns

    # discrete features (numerical with small unique values)
    discrete_features = [
        col for col in numerical_features
        if df[col].nunique() < 25
    ]

    # continuous features (numerical but not discrete)
    continuous_features = [
        col for col in numerical_features
        if col not in discrete_features
    ]

    print("Numerical Features:", len(numerical_features))
    print("Categorical Features:", len(categorical_features))
    print("Discrete Features:", len(discrete_features))
    print("Continuous Features:", len(continuous_features))

    return numerical_features, categorical_features, \
           discrete_features, discrete_features
           
res = feature_types(df)
```

```python
numerical_features,categorical_features,discrete_features,discrete_features = res
```

#### Test and Train split of the dataset

```python
# independent and dependent features
X = df.drop("prodtaken", axis=1) # independent feature
```

```python
y = df["prodtaken"]
```

```python
#test and train split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=24)
x_train.shape, x_test.shape
```

```python
numerical_features = [col for col in numerical_features if col != "prodtaken"]
numerical_features
```

#### Encoding

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

oh_transformer = OneHotEncoder(drop="first")
num_transformer = StandardScaler()

preprocessor = ColumnTransformer([
    ("OneHotEncoder", oh_transformer, categorical_features),
    ("StandardScaler", num_transformer, numerical_features)
])
```

```python
x_train = preprocessor.fit_transform(x_train)
```

```python
x_test = preprocessor.transform(x_test)
```

#### Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
y_prob = model.predict_proba(x_test)[:,1]
roc_auc_score(y_test, y_prob)
```
