```python
from sklearn.datasets import load_diabetes
d = load_diabetes()
```

```python
# separate the independent and dependent features
X = pd.DataFrame(d.data, columns=d.feature_names)
y = pd.Series(d.target, name="target")
```

```python
X
```

```python
# split the dataset into train and test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
x_train.head()
```

```python
x_train.corr()
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
```

```python
params = {
    "criterion":["squared_error"],
    "max_depth":[1,3,5,7],
    "min_samples_split":[10,20],
    "max_features":['sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
search = GridSearchCV(estimator=regressor,param_grid=params, cv=5, scoring="neg_mean_squared_error",refit=True, n_jobs=-1)
```

```python
search.fit(x_train, y_train)
```

```python
y_pred = search.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
